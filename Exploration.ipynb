#### Note: 

In this initial testing, we noticed a lot:

- Magnifi is a full financial website, not just prompt-and-response agent. The AI assitant was just one of many features on the website
- The platform uses Plaid, a company that allows fintech services to securely connect to user's 3rd-party accounts. We were honestly initially nervous about the pressure it was putting on us to link bank accounts, but we ended up putting in Kendall's Vanguard information to see how personalized it could be 
- The AI functionality had some visible flaws as a basic level. It was hard for to click the entry bar just to type a question (something was definitely faulty on the backend), and there was noticeably incorrect grammar
- It seems like the AI assistant runs mostly off of suggested, basic questions. Any time we asked something, it would direct us to ask the pre-loaded questions rather than continue with our own line of reasoning
- The platform was really user friendly, but lacked the depth/surety we expected from it. It asked us really easy-to-answer questions when we engaged with the retirement prompt, but then offered no timeline, tips, or ease for the portfolio development beyond being able to add the stocks to our cart via its purchase function
- Overall, it felt like Magnifi was made for people who already have a background in investing, know what they are doing, and just want to check specific stats or look at certain goals for which they are already planning. While the platform offers a lot for that demographic, our product will be much more entry-level and geared towards our target demographic of college students!
